```python
import naiveautoml
import sklearn.datasets
```

```python
X, y = sklearn.datasets.load_iris(return_X_y=True)
naml = naiveautoml.NaiveAutoML(show_progress=True,max_hpo_iterations=10)
```

```python
naml.fit(X, y)
print(naml.chosen_model)
```

```python
naml = naiveautoml.NaiveAutoML(show_progress=True)
naml.fit(X, y)
```

```python
naml.chosen_model
```

```python
naml.history
```

```python
naml.recover_model(history_index=25) # recover the model from the 25th iteration
```

Mediante la liberaria logging podemos ver que hace por detras la libreria naiveautoml:

```python
import logging
logger=logging.getLogger('naiveautoml')
logger.setLevel(logging.INFO)

ch=logging.StreamHandler()
ch.setLevel(logging.INFO)

formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
ch.setFormatter(formatter)
logger.addHandler(ch)
```

```python
naml=naiveautoml.NaiveAutoML(
    show_progress=True,
    scoring='accuracy',
    execution_timeout=10
    )
```

```python
naml.fit(X, y)
```
